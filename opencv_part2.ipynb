### 2.3 Modifikasi Simple Event Click

```python
windowName = "Original Image"

def callback_function(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("\rleft button pressed  ", end='')
    if event == cv2.EVENT_MOUSEMOVE:
        print("\rmouse moved          ", end='')
    if event == cv2.EVENT_LBUTTONUP:
        print("\rleft button released ", end='')

cv2.namedWindow(windowName)
cv2.setMouseCallback(windowName, callback_function)

image = cv2.imread('lena.jpg')

while True:
    cv2.imshow(windowName, image)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

### 2.4 Crop Image dengan Mouse Event Click

```python
x_start, y_start, x_end, y_end = 0, 0, 0, 0
windowName = "Original Image"

def crop_image(event, x, y, flags, param):
    global x_start, y_start, x_end, y_end

    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        x_end, y_end = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        croped_img = image[y_start:y_end, x_start:x_end]
        cv2.imshow("Cropped Image", croped_img)

cv2.namedWindow(windowName)
cv2.setMouseCallback(windowName, crop_image)

image = cv2.imread('lena.jpg')

while True:
    cv2.imshow(windowName, image)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

### 2.5 TASK: Crop + Save Image dengan Right Click

```python
x_start, y_start, x_end, y_end = 0, 0, 0, 0
windowName = "Original Image"

def crop_image(event, x, y, flags, param):
    global x_start, y_start, x_end, y_end

    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        x_end, y_end = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        croped_img = image[y_start:y_end, x_start:x_end]
        cv2.imshow("Cropped Image", croped_img)

    # TASK: Save image saat right click
    elif event == cv2.EVENT_RBUTTONDOWN:
        croped_img = image[y_start:y_end, x_start:x_end]
        cv2.imwrite("croped_image.png", croped_img)
        print("Image saved successfully as 'croped_image.png'!")

cv2.namedWindow(windowName)
cv2.setMouseCallback(windowName, crop_image)

image = cv2.imread('lena.jpg')

while True:
    cv2.imshow(windowName, image)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

### 3.6 Resize Interaktif dengan Mouse Event

```python
windowName = "Original Image"
is_resize = False

image = cv2.imread('lena.jpg')
h, w, c = image.shape
y_end, x_end = h, w

background = np.zeros_like(image)

def resize_image(event, x, y, flags, param):
    global x_end, y_end, is_resize

    if event == cv2.EVENT_LBUTTONDOWN:
        x_end, y_end = x, y
        is_resize = True

    elif event == cv2.EVENT_MOUSEMOVE and is_resize:
        x_end, y_end = x, y
        x_end = x_end if x_end < w else w
        y_end = y_end if y_end < h else h

    elif event == cv2.EVENT_LBUTTONUP:
        is_resize = False

cv2.namedWindow(windowName)
cv2.setMouseCallback(windowName, resize_image)

while True:
    template = background.copy()
    template[:y_end, :x_end] = cv2.resize(image, (x_end, y_end))

    cv2.imshow(windowName, template)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

### 4.1 Blending dua gambar ukuran sama

```python
alpha = 0.2

img1 = cv2.imread('lena.jpg')
img2 = cv2.imread('apple.jpg')

beta = 1.0 - alpha
blending_img = cv2.addWeighted(img1, alpha, img2, beta, 0.0)

cv2.imshow('Blending Result', blending_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4.2 Blending gambar beda ukuran - Stretch

```python
alpha = 0.5

img1 = cv2.imread('apple.jpg')
h, w, c = img1.shape

img2 = cv2.imread('pisang.jpg')
img2 = cv2.resize(img2, (w, h))  # stretch ke ukuran img1

beta = 1.0 - alpha
blending_img = cv2.addWeighted(img1, alpha, img2, beta, 0.0)

cv2.imshow('Blending Result (Stretch)', blending_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4.3 Blending gambar beda ukuran - Fit Window

```python
alpha = 0.3

img1 = cv2.imread('apple.jpg')
h1, w1, c1 = img1.shape

img2 = cv2.imread('pisang.jpg')
h2, w2, c2 = img2.shape

frame_img = np.ones_like(img1) * 255

if h2 > w2:
    h = h1
    w = int(w2 * h1 / h2)
else:
    w = w1
    h = int(h2 * w1 / w2)

frame_img[0:h, 0:w] = cv2.resize(img2, (w, h))

beta = 1.0 - alpha
blending_img = cv2.addWeighted(img1, alpha, frame_img, beta, 0.0)

cv2.imshow('Blending Result (Fit Window)', blending_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4.4 Overlay Transparent - Satu Objek

```python
alpha = 0.5

img = cv2.imread('lena.jpg')

overlay = np.zeros_like(img)
overlay[10:300, 10:50, 1] = 255  # kanal Green

img_blend = cv2.addWeighted(img, 1, overlay, alpha, 0.0)

cv2.imshow("Overlay Image (Green)", img_blend)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4.5 Overlay Transparent - Dua Objek

```python
alpha = 0.5

img = cv2.imread('lena.jpg')
overlay = np.zeros_like(img)

overlay_blue = overlay.copy()
overlay_blue[10:300, 10:50, 0] = 255  # kanal Blue
img_blend = cv2.addWeighted(img, 1, overlay_blue, alpha, 0.0)

overlay_red = overlay.copy()
overlay_red[10:100, 70:200, 2] = 255  # kanal Red
img_blend = cv2.addWeighted(img_blend, 1, overlay_red, alpha, 0.0)

cv2.imshow("Overlay Image (Blue + Red)", img_blend)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5. HOMEWORK: Overlay Biru Interaktif dengan Mouse Click

```python
windowName = "Blending Image"
is_drawing = False

image = cv2.imread('lena.jpg')
h, w, c = image.shape
x_start, y_start, x_end, y_end = 0, 0, 0, 0

background = np.zeros_like(image)

def draw_overlay(event, x, y, flags, param):
    global x_start, y_start, x_end, y_end, is_drawing

    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y
        is_drawing = True

    elif event == cv2.EVENT_MOUSEMOVE and is_drawing:
        x_end, y_end = x, y
        x_end = x_end if x_end < w else w
        y_end = y_end if y_end < h else h

    elif event == cv2.EVENT_LBUTTONUP:
        is_drawing = False

cv2.namedWindow(windowName)
cv2.setMouseCallback(windowName, draw_overlay)

while True:
    overlay_blue = background.copy()

    # Apply warna biru pada area yang dipilih (kanal index 0 = Blue di BGR)
    overlay_blue[y_start:y_end, x_start:x_end, 2] = 255

    # Blend gambar asli dengan overlay biru
    img_blend = cv2.addWeighted(image, 1, overlay_blue, 0.5, 0.0)

    cv2.imshow(windowName, img_blend)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```
